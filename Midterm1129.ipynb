```python
#Question 1a
mylist =[2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]
mylist.sort()
frequency = {}
for item in mylist:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1
for key,value in frequency.items():
    print(key,":",value)
    
```

```python
#question 1b
import matplotlib.pyplot as plt
import numpy as np
plt.hist(mylist,width=0.5)
plt.xlabel('Element')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
#question 1c
import json
y = json.dumps(frequency)
print(y)

f = open("Midterm1129","w")
f.write(y)
f.close()

f=open("Midterm1129","r")
print(f.read())
```

```python
#question 2
import pandas as pd

df = pd.read_csv('amazon-orders.csv')
df.head()
```

```python

df.shape
```

```python
df = df.fillna(0) 
df.head()
```

```python
df["Item Total"] = df["Item Total"].str.replace('$','').astype(float)
df.head()
```

```python
df["Item Total"].sum()
```

```python
df["Item Subtotal Tax"] = df["Item Subtotal Tax"].str.replace('$','').astype(float)
df.head()
```

```python
df["Item Subtotal Tax"].sum()
```

```python
df["Item Total"].sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
my_orders=df.groupby('Order Date').sum()["Item Total"]
my_orders.head()
```

```python
%matplotlib inline
df.plot.bar(x="Order Date",y="Item Total",title="Item Total per order date",rot=90,figsize=(20,10),ylabel="Amount")
```

```python
my_monthly_orders=df.groupby(pd.Grouper(key='Order Date',freq='1M')).sum()["Item Total"]
my_monthly_orders.head()
```

```python
my_monthly_orders.plot.bar(figsize=(20,10),title='My_monthly_oders',color='#61D199',ylabel='Amount')
```
