```r
mydata<- read.csv('C:/Users/11840/myorder.csv')
head(mydata)
dim(mydata)
mydata[is.na(mydata)] <- 0 
head(mydata)
```

```r
#total
mydata$Item.Total <- as.numeric(gsub("\\$", "", mydata$Item.Total))
mydata$Item.Total
```

```r
min(mydata$Item.Total)
max(mydata$Item.Total)
median(mydata$Item.Total)
sum(mydata$Item.Total)
mean(mydata$Item.Total)
```

```r
#tax
mydata$Item.Subtotal.Tax <- as.numeric(gsub("\\$", "", mydata$Item.Subtotal.Tax))
mydata$Item.Subtotal.Tax
min(mydata$Item.Subtotal.Tax)
max(mydata$Item.Subtotal.Tax)
median(mydata$Item.Subtotal.Tax)
sum(mydata$Item.Subtotal.Tax)
mean(mydata$Item.Subtotal.Tax)
```

```r
#tax rate
sum(mydata$Item.Subtotal.Tax) / sum(mydata$Item.Total)
```

```r
#tax plot
barplot(mydata$Item.Subtotal.Tax, names.arg=mydata$Order.Date,col="blue",main = "Item Subtotal Tax in each Order Date", xlab = " The Order Date", ylab = "Item Subtotal Tax")
```

```r
#item total charged plot
barplot(mydata$Item.Total, names.arg=mydata$Order.Date,col="red",main = "Item.Total in each Order Date", xlab = " The Order Date", ylab = "Item.Total")

```
