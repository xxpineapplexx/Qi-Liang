```r
#Question2: change to for loop using R

install.packages("TurtleGraphics")
library("TurtleGraphics")
turtle_init()
rectangle<-function(length,height,color){
  for (i in 1:2){
    turtle_forward(length)
    turtle_right(90)
    turtle_forward(height)
    turtle_right(90)
  }
}
rectangle(10,20,"Old Glory Red")
```

```r
#for loop original 
def draw_rectangle(length,height,color):
    turtle.speed(100)           #set speed for drawing
    turtle.fillcolor(color)     #set color, parameter defined below
    turtle.begin_fill()         #beginning of the drawing
    for i in range(2):          #repeat two times the following steps to draw a rectangle
        turtle.forward(length)
        turtle.right(90)
        turtle.forward(height)
        turtle.right(90)
    turtle.end_fill()          
```
